```python
import pandas as pd

df = pd.read_csv("Data.csv")

print(df.head())
```

```python
print(df.info())
print(df.shape)
```

```python
# Total EV sales
total_sales = df["EV_Sales_Quantity"].sum()
print("Total EV sales:", total_sales)

# Sales by year
sales_year = df.groupby("Year")["EV_Sales_Quantity"].sum()
print(sales_year)

# Top states
sales_state = df.groupby("State")["EV_Sales_Quantity"].sum().sort_values(ascending=False)
print(sales_state.head(10))
```

```python
import matplotlib.pyplot as plt

sales_year.plot()
plt.title("EV Sales Trend")
plt.show()
```

```python
import pandas as pd
df = pd.read_csv("Data.csv")
print(df.head())
```

```python
print(df.info())
print(df.shape)
print(df.columns)
```

```python
df["Year"] = df["Year"].astype(int)
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")

print(df.isnull().sum())
print(df.duplicated().sum())
```

```python
total_sales = df["EV_Sales_Quantity"].sum()
print("Total EV Sales in India:", total_sales)
```

```python
sales_year = df.groupby("Year")["EV_Sales_Quantity"].sum()
print(sales_year)
```

```python
sales_state = df.groupby("State")["EV_Sales_Quantity"].sum().sort_values(ascending=False)
print(sales_state.head(10))
```

```python
sales_cat = df.groupby("Vehicle_Category")["EV_Sales_Quantity"].sum()
print(sales_cat)
```
